```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
x=np.linspace(-10,10)
plt.plot(x,sigmoid(x))
plt.title('Activation function sigmoid')
plt.show
```
